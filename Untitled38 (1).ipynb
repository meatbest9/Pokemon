```python
import ipywidgets as widgets
from IPython.display import display, HTML
import random
import time

# ==========================
# 0. 데이터: 타입 상성
# ==========================
TYPE_EFFECTIVENESS = {
    '노말': {'super_effective': [], 'not_very_effective': ['바위', '강철'], 'immune': ['고스트']},
    '불꽃': {'super_effective': ['풀', '얼음', '벌레', '강철'], 'not_very_effective': ['불꽃', '물', '바위', '드래곤'], 'immune': []},
    '물': {'super_effective': ['불꽃', '땅', '바위'], 'not_very_effective': ['물', '풀', '드래곤'], 'immune': []},
    '전기': {'super_effective': ['물', '비행'], 'not_very_effective': ['전기', '풀', '드래곤'], 'immune': ['땅']},
    '풀': {'super_effective': ['물', '땅', '바위'], 'not_very_effective': ['불꽃', '풀', '독', '비행', '벌레', '드래곤', '강철'], 'immune': []},
    '얼음': {'super_effective': ['풀', '땅', '비행', '드래곤'], 'not_very_effective': ['불꽃', '물', '얼음', '강철'], 'immune': []},
    '격투': {'super_effective': ['노말', '얼음', '바위', '악', '강철'], 'not_very_effective': ['독', '비행', '에스퍼', '벌레'], 'immune': ['고스트']},
    '독': {'super_effective': ['풀'], 'not_very_effective': ['독', '땅', '바위', '고스트'], 'immune': ['강철']},
    '땅': {'super_effective': ['불꽃', '전기', '독', '바위', '강철'], 'not_very_effective': ['풀', '벌레'], 'immune': ['비행']},
    '비행': {'super_effective': ['풀', '격투', '벌레'], 'not_very_effective': ['전기', '바위', '강철'], 'immune': []},
    '에스퍼': {'super_effective': ['격투', '독'], 'not_very_effective': ['에스퍼', '강철'], 'immune': ['악']},
    '벌레': {'super_effective': ['풀', '에스퍼', '악'], 'not_very_effective': ['불꽃', '격투', '독', '비행', '고스트', '강철'], 'immune': []},
    '바위': {'super_effective': ['불꽃', '얼음', '비행', '벌레'], 'not_very_effective': ['격투', '땅', '강철'], 'immune': []},
    '고스트': {'super_effective': ['에스퍼', '고스트'], 'not_very_effective': ['악', '강철'], 'immune': ['노말']},
    '드래곤': {'super_effective': ['드래곤'], 'not_very_effective': ['강철'], 'immune': []},
    '악': {'super_effective': ['에스퍼', '고스트'], 'not_very_effective': ['격투', '악', '강철'], 'immune': []},
    '강철': {'super_effective': ['얼음', '바위'], 'not_very_effective': ['불꽃', '물', '전기', '강철'], 'immune': []}
}

def calculate_effectiveness(attack_type, defend_type):
    if attack_type not in TYPE_EFFECTIVENESS:
        return 1.0
    chart = TYPE_EFFECTIVENESS[attack_type]
    if defend_type in chart['immune']:
        return 0.0
    if defend_type in chart['super_effective']:
        return 2.0
    if defend_type in chart['not_very_effective']:
        return 0.5
    return 1.0

# ==========================
# 1. 포켓몬 데이터 (야생 16종)
# ==========================
WILD_SPECIES_DATA = [
    {"name": "리자몽", "types": ["불꽃", "비행"], "base": [78, 84, 78, 109, 85, 100],
     "moves": [
         {"name": "화염방사", "power": 95, "type": "불꽃", "category": "special"},
         {"name": "에어슬래시", "power": 75, "type": "비행", "category": "special"},
         {"name": "용의파동", "power": 85, "type": "드래곤", "category": "special"},
         {"name": "솔라빔", "power": 120, "type": "풀", "category": "special", "charge_turn": True}
     ]},
    {"name": "갸라도스", "types": ["물", "비행"], "base": [95, 125, 79, 60, 100, 81],
     "moves": [
         {"name": "폭포오르기", "power": 80, "type": "물", "category": "physical"},
         {"name": "지진", "power": 100, "type": "땅", "category": "physical"},
         {"name": "얼음엄니", "power": 65, "type": "얼음", "category": "physical"},
         {"name": "용의춤", "power": 0, "type": "드래곤", "category": "status"}
     ]},
    {"name": "라이코", "types": ["전기"], "base": [90, 85, 75, 115, 100, 115],
     "moves": [
         {"name": "10만볼트", "power": 95, "type": "전기", "category": "special"},
         {"name": "볼트체인지", "power": 70, "type": "전기", "category": "special"},
         {"name": "잠재파워-얼음", "power": 60, "type": "얼음", "category": "special"},
         {"name": "섀도볼", "power": 80, "type": "고스트", "category": "special"}
     ]},
    {"name": "로즈레이드", "types": ["풀", "독"], "base": [60, 70, 65, 125, 105, 90],
     "moves": [
         {"name": "리프스톰", "power": 130, "type": "풀", "category": "special"},
         {"name": "오물폭탄", "power": 90, "type": "독", "category": "special"},
         {"name": "섀도볼", "power": 80, "type": "고스트", "category": "special"},
         {"name": "잠재파워-불꽃", "power": 60, "type": "불꽃", "category": "special"}
     ]},
    {"name": "포푸니라", "types": ["악", "얼음"], "base": [70, 120, 65, 45, 85, 125],
     "moves": [
         {"name": "고드름떨구기", "power": 85, "type": "얼음", "category": "physical"},
         {"name": "속이다", "power": 40, "type": "노말", "category": "physical", "priority": 1},
         {"name": "기습", "power": 40, "type": "악", "category": "physical", "priority": 1},
         {"name": "냉동펀치", "power": 75, "type": "얼음", "category": "physical"}
     ]},
    {"name": "루카리오", "types": ["격투", "강철"], "base": [70, 110, 70, 115, 70, 90],
     "moves": [
         {"name": "인파이트", "power": 120, "type": "격투", "category": "physical"},
         {"name": "신속", "power": 40, "type": "노말", "category": "physical", "priority": 1},
         {"name": "파동탄", "power": 80, "type": "격투", "category": "special", "accuracy": 1000},
         {"name": "스톤에지", "power": 100, "type": "바위", "category": "physical"}
     ]},
    {"name": "드래피온", "types": ["독", "악"], "base": [70, 90, 110, 60, 75, 95],
     "moves": [
         {"name": "시저크로스", "power": 80, "type": "벌레", "category": "physical"},
         {"name": "지진", "power": 100, "type": "땅", "category": "physical"},
         {"name": "독찌르기", "power": 80, "type": "독", "category": "physical"},
         {"name": "바위봉인", "power": 60, "type": "바위", "category": "physical"}
     ]},
    {"name": "한카리아스", "types": ["드래곤", "땅"], "base": [108, 130, 95, 80, 85, 102],
     "moves": [
         {"name": "지진", "power": 100, "type": "땅", "category": "physical"},
         {"name": "역린", "power": 120, "type": "드래곤", "category": "physical"},
         {"name": "스톤에지", "power": 100, "type": "바위", "category": "physical"},
         {"name": "칼춤", "power": 0, "type": "노말", "category": "status"}
     ]},
    {"name": "마기라스", "types": ["바위", "악"], "base": [100, 134, 110, 95, 100, 61],
     "moves": [
         {"name": "스톤에지", "power": 100, "type": "바위", "category": "physical"},
         {"name": "깨물어부수기", "power": 80, "type": "악", "category": "physical"},
         {"name": "지진", "power": 100, "type": "땅", "category": "physical"},
         {"name": "용의춤", "power": 0, "type": "드래곤", "category": "status"}
     ]},
    {"name": "토게키스", "types": ["노말", "비행"], "base": [85, 50, 95, 120, 115, 80],
     "moves": [
         {"name": "에어슬래시", "power": 75, "type": "비행", "category": "special", "effect": "flinch", "effect_chance": 60},
         {"name": "파동탄", "power": 80, "type": "격투", "category": "special"},
         {"name": "화염방사", "power": 95, "type": "불꽃", "category": "special"},
         {"name": "매지컬샤인", "power": 80, "type": "노말", "category": "special"}
     ]},
    {"name": "후딘", "types": ["에스퍼"], "base": [55, 50, 45, 135, 95, 120],
     "moves": [
         {"name": "사이코키네시스", "power": 90, "type": "에스퍼", "category": "special"},
         {"name": "섀도볼", "power": 80, "type": "고스트", "category": "special"},
         {"name": "기합구슬", "power": 120, "type": "격투", "category": "special"},
         {"name": "에너지볼", "power": 90, "type": "풀", "category": "special"}
     ]},
    {"name": "핫삼", "types": ["벌레", "강철"], "base": [70, 130, 100, 55, 80, 65],
     "moves": [
         {"name": "불릿펀치", "power": 40, "type": "강철", "category": "physical", "priority": 1},
         {"name": "유턴", "power": 70, "type": "벌레", "category": "physical"},
         {"name": "시저크로스", "power": 80, "type": "벌레", "category": "physical"},
         {"name": "칼춤", "power": 0, "type": "노말", "category": "status"}
     ]},
    {"name": "팬텀", "types": ["고스트", "독"], "base": [60, 65, 60, 130, 75, 110],
     "moves": [
         {"name": "섀도볼", "power": 80, "type": "고스트", "category": "special"},
         {"name": "오물폭탄", "power": 90, "type": "독", "category": "special"},
         {"name": "기합구슬", "power": 120, "type": "격투", "category": "special"},
         {"name": "10만볼트", "power": 95, "type": "전기", "category": "special"}
     ]},
    {"name": "보만다", "types": ["드래곤", "비행"], "base": [95, 135, 80, 110, 80, 100],
     "moves": [
         {"name": "역린", "power": 120, "type": "드래곤", "category": "physical"},
         {"name": "용성군", "power": 130, "type": "드래곤", "category": "special"},
         {"name": "지진", "power": 100, "type": "땅", "category": "physical"},
         {"name": "용의춤", "power": 0, "type": "드래곤", "category": "status"}
     ]},
    {"name": "절각참", "types": ["악", "강철"], "base": [65, 125, 100, 60, 70, 70],
     "moves": [
         {"name": "기습", "power": 70, "type": "악", "category": "physical", "priority": 1},
         {"name": "아이언헤드", "power": 80, "type": "강철", "category": "physical"},
         {"name": "칼춤", "power": 0, "type": "노말", "category": "status"},
         {"name": "스톤에지", "power": 100, "type": "바위", "category": "physical"}
     ]},
    {"name": "메타그로스", "types": ["강철", "에스퍼"], "base": [80, 135, 130, 95, 90, 70],
     "moves": [
         {"name": "코멧펀치", "power": 90, "type": "강철", "category": "physical"},
         {"name": "지진", "power": 100, "type": "땅", "category": "physical"},
         {"name": "사념의박치기", "power": 80, "type": "에스퍼", "category": "physical"},
         {"name": "불릿펀치", "power": 40, "type": "강철", "category": "physical", "priority": 1}
     ]}
]

# ==========================
# 2. 플레이어 전용 포켓몬 데이터 (6마리)
# ==========================
PLAYER_PARTY_DATA = [
    {
        "name": "리자몽", "types": ["불꽃", "비행"], "base": [78, 84, 78, 109, 85, 100],
        "moves": [
            {"name": "화염방사", "power": 95, "type": "불꽃", "category": "special", "effect": "burn", "effect_chance": 10},
            {"name": "에어슬래시", "power": 75, "type": "비행", "category": "special", "effect": "flinch", "effect_chance": 30},
            {"name": "용의파동", "power": 85, "type": "드래곤", "category": "special"},
            {"name": "솔라빔", "power": 120, "type": "풀", "category": "special", "charge_turn": True}
        ]
    },
    {
        "name": "갸라도스", "types": ["물", "비행"], "base": [95, 125, 79, 60, 100, 81],
        "moves": [
            {"name": "폭포오르기", "power": 80, "type": "물", "category": "physical", "effect": "flinch", "effect_chance": 20},
            {"name": "지진", "power": 100, "type": "땅", "category": "physical"},
            {"name": "얼음엄니", "power": 65, "type": "얼음", "category": "physical", "effect": "freeze", "effect_chance": 10},
            {"name": "용의춤", "power": 0, "type": "드래곤", "category": "status"} 
        ]
    },
    {
        "name": "라이코", "types": ["전기"], "base": [90, 85, 75, 115, 100, 115],
        "moves": [
            {"name": "10만볼트", "power": 95, "type": "전기", "category": "special", "effect": "paralysis", "effect_chance": 10},
            {"name": "볼트체인지", "power": 70, "type": "전기", "category": "special"},
            {"name": "잠재파워", "power": 60, "type": "얼음", "category": "special"},
            {"name": "섀도볼", "power": 80, "type": "고스트", "category": "special"}
        ]
    },
    {
        "name": "로즈레이드", "types": ["풀", "독"], "base": [60, 70, 65, 125, 105, 90],
        "moves": [
            {"name": "리프스톰", "power": 130, "type": "풀", "category": "special"},
            {"name": "슬러지봄", "power": 90, "type": "독", "category": "special", "effect": "poison", "effect_chance": 30},
            {"name": "섀도볼", "power": 80, "type": "고스트", "category": "special"},
            {"name": "잠재파워", "power": 60, "type": "불꽃", "category": "special"}
        ]
    },
    {
        "name": "포푸니라", "types": ["악", "얼음"], "base": [70, 120, 65, 45, 85, 125],
        "moves": [
            {"name": "고드름떨구기", "power": 85, "type": "얼음", "category": "physical"},
            {"name": "속이다", "power": 40, "type": "노말", "category": "physical", "priority": 1},
            {"name": "기습", "power": 40, "type": "악", "category": "physical", "priority": 1},
            {"name": "냉동펀치", "power": 75, "type": "얼음", "category": "physical", "effect": "freeze", "effect_chance": 10}
        ]
    },
    {
        "name": "루카리오", "types": ["격투", "강철"], "base": [70, 110, 70, 115, 70, 90],
        "moves": [
            {"name": "인파이트", "power": 120, "type": "격투", "category": "physical"},
            {"name": "신속", "power": 40, "type": "노말", "category": "physical", "priority": 1},
            {"name": "파동탄", "power": 80, "type": "격투", "category": "special", "accuracy": 1000},
            {"name": "스톤에지", "power": 100, "type": "바위", "category": "physical", "accuracy": 80}
        ]
    }
]

# ------------------------------------------------------------------
# 3. 클래스 정의
# ------------------------------------------------------------------
class Pokemon:
    def __init__(
        self, name, level, max_hp, attack, defense,
        sp_attack, sp_defense, speed, moves, types
    ):
        self.name = name
        self.level = level
        self.max_hp = max_hp
        self.current_hp = max_hp
        self.attack = attack
        self.defense = defense
        self.sp_attack = sp_attack
        self.sp_defense = sp_defense
        self.speed = speed
        self.moves = moves
        self.types = types
        self.is_fainted = False
        self.is_charging = None

        self.status = None
        self.status_counter = 0
        self.is_confused = False
        self.confused_counter = 0

        # 💥 이번 턴에 데미지 받았는지 (렌더에서 흔들림용)
        self.just_took_damage = False

    def check_can_move(self):
        if self.status == '얼음':
            if random.random() < 0.2:
                print(f"▶ {self.name}의 얼음이 녹았다!")
                self.status = None
            else:
                print(f"▶ {self.name}은(는) 얼어서 움직일 수 없다!")
                return False

        if self.status == '수면':
            self.status_counter -= 1
            if self.status_counter <= 0:
                print(f"▶ {self.name}이(가) 잠에서 깨어났다!")
                self.status = None
            else:
                print(f"▶ {self.name}은(는) 쿨쿨 잠들어 있다...")
                return False

        if self.status == '마비':
            if random.random() < 0.25:
                print(f"▶ {self.name}은(는) 몸이 저려서 움직일 수 없다!")
                return False

        if self.is_confused:
            print(f"▶ {self.name}은(는) 혼란에 빠져있다!")
            self.confused_counter -= 1
            if self.confused_counter <= 0:
                self.is_confused = False
                print(f"▶ {self.name}의 혼란이 풀렸다!")
            else:
                if random.random() < 0.5:
                    print(f"▶ {self.name}은(는) 혼란으로 자신을 공격했다!")
                    damage = int(((self.level * 0.4 + 2) * self.attack * 40) / (self.defense * 50) + 2)
                    self.take_damage(damage)
                    return False
        return True

    def attack_opponent(self, opponent, move, battle_weather):
        if not self.check_can_move():
            return

        if move.get('charge_turn') and self.is_charging is None:
            if battle_weather == '쾌청':
                print(f"▶ 쾌청 날씨로 {move['name']} 즉시 발동!")
            else:
                self.is_charging = move
                print(f"▶ {self.name}이(가) 빛을 모으기 시작했다!")
                return

        if self.is_charging:
            print(f"▶ {self.name}의 {self.is_charging['name']} 발사!")
            move = self.is_charging
            self.is_charging = None
        else:
            print(f"▶ {self.name}이(가) {move['name']} 사용!")
            time.sleep(0.2)

        move_accuracy = move.get('accuracy', 100)
        if move['name'] == '번개' and battle_weather == '비':
            move_accuracy = 100

        if random.randint(1, 100) > move_accuracy:
            print("...하지만 빗나갔다!")
            return

        if move['category'] == 'status':
            self.apply_status_effect(opponent, move)
            return

        stab = 1.5 if move['type'] in self.types else 1.0
        total_effectiveness = 1.0
        for opponent_type in opponent.types:
            total_effectiveness *= calculate_effectiveness(move['type'], opponent_type)

        weather_mod = 1.0
        if battle_weather == '비':
            if move['type'] == '물':
                weather_mod = 1.5
            elif move['type'] == '불꽃':
                weather_mod = 0.5
        elif battle_weather == '쾌청':
            if move['type'] == '불꽃':
                weather_mod = 1.5
            elif move['type'] == '물':
                weather_mod = 0.5

        burn_mod = 0.5 if (self.status == '화상' and move['category'] == 'physical') else 1.0

        if move['category'] == 'physical':
            attacker_stat = self.attack
            defender_stat = opponent.defense
        else:
            attacker_stat = self.sp_attack
            defender_stat = opponent.sp_defense

        if battle_weather == '모래바람' and '바위' in opponent.types:
            defender_stat = int(defender_stat * 1.5)

        base_damage = int(((self.level * 0.4 + 2) * attacker_stat * move['power']) / (defender_stat * 50) + 2)
        random_mod = random.uniform(0.85, 1.0)
        final_damage = int(base_damage * stab * total_effectiveness * weather_mod * burn_mod * random_mod)

        if final_damage <= 0 and total_effectiveness > 0:
            final_damage = 1

        if total_effectiveness == 0.0:
            print("▶ ...효과가 없는 것 같다.")
        elif total_effectiveness > 1.0:
            print("▶ 효과는 굉장했다!")
        elif total_effectiveness < 1.0:
            print("▶ 효과가 별로인 듯하다...")

        print(f"▶ {opponent.name}에게 {final_damage}의 데미지!")
        opponent.take_damage(final_damage)

        if 'effect' in move and random.randint(1, 100) <= move.get('effect_chance', 0):
            self.apply_status_effect(opponent, {'effect': move['effect']})

    def apply_status_effect(self, target, move):
        effect = move.get('effect')
        if not effect:
            return

        if effect == 'sleep':
            if target.status:
                print(f"▶ {target.name}은(는) 이미 상태 이상에 걸려있다.")
            else:
                target.status = '수면'
                target.status_counter = random.randint(1, 3)
                print(f"▶ {target.name}은(는) 잠들어버렸다!")
        elif effect == 'poison':
            if target.status:
                print(f"▶ {target.name}은(는) 이미 상태 이상에 걸려있다.")
            elif '독' in target.types or '강철' in target.types:
                print("▶ 독이 통하지 않는다!")
            else:
                target.status = '독'
                print(f"▶ {target.name}은(는) 독에 걸렸다!")
        elif effect == 'bad_poison':
            if target.status:
                print(f"▶ {target.name}은(는) 이미 상태 이상에 걸려있다.")
            elif '독' in target.types or '강철' in target.types:
                print("▶ 독이 통하지 않는다!")
            else:
                target.status = '맹독'
                target.status_counter = 0
                print(f"▶ {target.name}은(는) 맹독에 걸렸다!")
        elif effect == 'burn':
            if target.status:
                print(f"▶ {target.name}은(는) 이미 상태 이상에 걸려있다.")
            elif '불꽃' in target.types:
                print("▶ 불꽃 타입에게는 화상이 통하지 않는다!")
            else:
                target.status = '화상'
                print(f"▶ {target.name}은(는) 화상을 입었다!")
        elif effect == 'paralysis':
            if target.status:
                print(f"▶ {target.name}은(는) 이미 상태 이상에 걸려있다.")
            elif '전기' in target.types:
                print("▶ 전기 타입에게는 마비가 통하지 않는다!")
            else:
                target.status = '마비'
                print(f"▶ {target.name}은(는) 마비되어 기술이 나오기 힘들어졌다!")
        elif effect == 'freeze':
            if target.status:
                print(f"▶ {target.name}은(는) 이미 상태 이상에 걸려있다.")
            elif '얼음' in target.types:
                print("▶ 얼음 타입은 얼지 않는다!")
            else:
                target.status = '얼음'
                print(f"▶ {target.name}은(는) 꽁꽁 얼어버렸다!")
        elif effect == 'confusion':
            if target.is_confused:
                print(f"▶ {target.name}은(는) 이미 혼란에 빠져있다.")
            else:
                target.is_confused = True
                target.confused_counter = random.randint(1, 4)
                print(f"▶ {target.name}은(는) 혼란에 빠졌다!")

    def take_damage(self, damage):
        # 💥 0 이하면 무시 (흔들리지도 않음)
        if damage <= 0:
            return

        # 💥 이번 턴에 데미지 받음 → 렌더에서 흔들리게 표시
        self.just_took_damage = True

        self.current_hp -= damage
        if self.current_hp <= 0:
            self.current_hp = 0
            self.is_fainted = True
            self.status = None
            self.is_confused = False
            print(f"▶ {self.name}이(가) 쓰러졌다!")
        else:
            print(f"▶ {self.name}의 남은 HP: {self.current_hp}/{self.max_hp}")

    def heal(self, amount):
        if self.is_fainted:
            print(f"▶ {self.name}은(는) 기절해서 회복할 수 없다!")
            return
        self.current_hp += amount
        if self.current_hp > self.max_hp:
            self.current_hp = self.max_hp
        print(f"▶ {self.name}의 HP 회복! ({self.current_hp}/{self.max_hp})")

    def apply_end_of_turn_status_damage(self):
        if self.is_fainted:
            return
        damage = 0
        msg = ""
        if self.status == '독':
            damage = self.max_hp // 16
            msg = "독으로 인해 고통받고 있다!"
        elif self.status == '맹독':
            self.status_counter += 1
            damage = (self.max_hp // 16) * self.status_counter
            msg = "맹독이 퍼져나간다!"
        elif self.status == '화상':
            damage = self.max_hp // 16
            msg = "화상을 입어 괴롭다!"

        if damage > 0:
            print(f"▶ {self.name}은(는) {msg}")
            self.take_damage(damage)

class Trainer:
    def __init__(self, name, team, items):
        self.name = name
        self.team = team
        self.items = items
        self.current_pokemon = team[0]

    def switch_pokemon(self, new_pokemon_index):
        if new_pokemon_index < 0 or new_pokemon_index >= len(self.team):
            return False
        new_pokemon = self.team[new_pokemon_index]
        if new_pokemon.is_fainted:
            return False
        if new_pokemon == self.current_pokemon:
            return False

        if self.current_pokemon.is_confused:
            self.current_pokemon.is_confused = False
        if self.current_pokemon.status == '맹독':
            self.current_pokemon.status_counter = 0

        print(f"▶ 돌아와, {self.current_pokemon.name}!")
        self.current_pokemon = new_pokemon
        print(f"▶ 가라, {self.current_pokemon.name}!")
        return True

    def use_item(self, item_name, target_pokemon):
        # 1. 아이템 보유량 체크
        if item_name not in self.items or self.items[item_name] <= 0:
            print("▶ 아이템이 없습니다.")
            return False

        # 2. 아이템 소모 (배틀 인벤토리)
        self.items[item_name] -= 1
        print(f"▶ {item_name} 사용!")

        # 3. [중요] 전역 가방(state['bag']) 동기화
        global state
        found = False
        for category in state['bag']:
            for item in state['bag'][category]:
                if item['name'] == item_name:
                    item['count'] -= 1
                    found = True
                    break
            if found:
                break

        # 4. 효과 적용
        if item_name == '상처약':
            target_pokemon.heal(20)
        elif item_name == '좋은상처약':
            target_pokemon.heal(50)
        elif item_name == '고급상처약':
            target_pokemon.heal(200)
        elif item_name == '맛있는 물':
            target_pokemon.heal(30)
        elif item_name == '만병통치제':
            target_pokemon.status = None
            target_pokemon.is_confused = False
            print(f"▶ {target_pokemon.name}의 상태 이상이 모두 회복되었다!")

        return True

    def all_pokemon_fainted(self):
        return all(p.is_fainted for p in self.team)

    def get_ai_choice(self):
        if self.current_pokemon.is_charging:
            return {'action': 'attack', 'priority': 0, 'move': self.current_pokemon.is_charging}
        move = random.choice(self.current_pokemon.moves)
        return {'action': 'attack', 'priority': 0, 'move': move}

# ------------------------------------------------------------------
# 4. 헬퍼 함수: 포켓몬 생성
# ------------------------------------------------------------------
def create_pokemon_from_data(data, level):
    max_hp = int((data["base"][0] * 2 * level) / 100 + level + 10)
    stats = []
    for base in data["base"][1:]:
        val = int((base * 2 * level) / 100 + 5)
        stats.append(val)
    atk, dfs, spa, spd, spe = stats

    return Pokemon(
        name=data["name"],
        level=level,
        max_hp=max_hp,
        attack=atk,
        defense=dfs,
        sp_attack=spa,
        sp_defense=spd,
        speed=spe,
        moves=data["moves"],
        types=data["types"]
    )

def get_random_wild_pokemon(level=50):
    data = random.choice(WILD_SPECIES_DATA)
    return create_pokemon_from_data(data, level)

# ------------------------------------------------------------------
# 5. UI 배틀 전역 상태
# ------------------------------------------------------------------
ui_battle = {
    "active": False,
    "player": None,
    "enemy": None,
    "weather": None,
    "weather_turns": 0,
    "turn": 1,
}

# ==================================================================
# =======================  UI 코드  ======================
# ==================================================================

SCALE = 1.2
BASE_SCREEN_WIDTH = 800
BASE_SCREEN_HEIGHT = 260
BASE_PANEL_WIDTH = 380
BASE_BATTLE_HEIGHT = 200
BASE_MSG_HEIGHT = 60

SCREEN_WIDTH = int(BASE_SCREEN_WIDTH * SCALE)
SCREEN_HEIGHT = int(BASE_SCREEN_HEIGHT * SCALE)
PANEL_WIDTH = SCREEN_WIDTH // 2
BATTLE_HEIGHT = int(BASE_BATTLE_HEIGHT * SCALE)
MSG_HEIGHT = SCREEN_HEIGHT - BATTLE_HEIGHT

FIGHT_HEIGHT = int(120 * SCALE)
BTN_W_SMALL = int(110 * SCALE)
BTN_H_SMALL = int(60 * SCALE)
ITEM_W = int(145 * SCALE)
ITEM_H = int(50 * SCALE)
CAT_W = int(300 * SCALE)
CAT_H = int(50 * SCALE)
BACK_H = int(40 * SCALE)
NAV_W = int(70 * SCALE)

button_style = HTML("""
<style>
.poke-btn {
    border: 2px solid #000 !important;
    border-radius: 8px !important;
    font-weight: bold !important;
    color: #ffffff !important;
    font-size: 20px !important;
    text-align: center !important;
    text-shadow: -1px 0 #000, 1px 0 #000, 0 -1px #000, 0 1px #000,
                 -1px -1px #000, 1px -1px #000, -1px 1px #000, 1px 1px #000;
}
.poke-red   { background: linear-gradient(to bottom, #ffbcbc 0%, #ffbcbc 50%, #ff8282 55%, #d35050 100%) !important; }
.poke-brown { background: linear-gradient(to bottom, #ffee96 0%, #ffee96 50%, #fcca5b 55%, #b8892c 100%) !important; }
.poke-blue  { background: linear-gradient(to bottom, #c6e1ff 0%, #c6e1ff 50%, #8cbcff 55%, #2c5fa8 100%) !important; }
.poke-green { background: linear-gradient(to bottom, #c9f8b0 0%, #c9f8b0 50%, #89df63 55%, #3c8d25 100%) !important; }

/* 💥 맞았을 때 흔들리는 애니메이션 */
@keyframes hit-shake {
    0%   { transform: translateX(0); }
    20%  { transform: translateX(-15px); }
    40%  { transform: translateX(15px); }
    60%  { transform: translateX(-15px); }
    80%  { transform: translateX(15px); }
    100% { transform: translateX(0); }
}

.shake {
    animation: hit-shake 0.25s linear 1;
}
</style>
""")
display(button_style)

debug_out = widgets.Output()

state = {
    "screen": "MAIN_MENU",
    "message": "지우는 무엇을 할까?",
    "bag_category": None,
    "bag_page": 0,
    # 메인 메뉴용 더미 데이터 (실제 배틀은 PLAYER_PARTY_DATA 사용)
    "party": [{"name": p["name"], "level": 50, "hp": 100, "max_hp": 100} for p in PLAYER_PARTY_DATA],
    "bag": {
        "HP_PP": [
            {"name": "상처약", "count": 2}, {"name": "PP에이드", "count": 1},
            {"name": "고급상처약", "count": 1}, {"name": "좋은상처약", "count": 1},
            {"name": "맛있는 물", "count": 6}, {"name": "PP회복", "count": 1},
            {"name": "고급좋은상처약", "count": 4}
        ],
        "BALL": [{"name": "몬스터볼", "count": 5}, {"name": "수퍼볼", "count": 3}],
        "STATUS": [{"name": "해독제", "count": 2}, {"name": "화상치료제", "count": 1}],
        "BATTLE": [{"name": "공격업", "count": 1}, {"name": "스피드업", "count": 1}]
    }
}

# ---------------- UI 초기화 함수 ----------------
def init_ui_battle():
    if ui_battle["player"] is not None:
        return

    # 1. 플레이어 파티 생성
    player_team = []
    for p_data in PLAYER_PARTY_DATA:
        new_mon = create_pokemon_from_data(p_data, 50)
        player_team.append(new_mon)

    # 2. 현재 가방 상태를 가져와서 트레이너에게 지급
    battle_items = {}
    for cat in state['bag']:
        for item in state['bag'][cat]:
            if item['count'] > 0:
                battle_items[item['name']] = item['count']

    player = Trainer("지우", player_team, battle_items)
    
    dummy_mon = get_random_wild_pokemon(50)
    rival = Trainer("야생", [dummy_mon], {})

    ui_battle["player"] = player
    ui_battle["enemy"] = rival
    ui_battle["active"] = False
    ui_battle["weather"] = None
    ui_battle["turn"] = 1

# ---------------- 배틀 시작 (랜덤 포켓몬 생성) ----------------
def start_ui_battle():
    init_ui_battle()
    
    wild_level = random.randint(45, 55)
    wild_mon = get_random_wild_pokemon(wild_level)
    
    ui_battle["enemy"] = Trainer("야생", [wild_mon], {})
    ui_battle["active"] = True
    ui_battle["turn"] = 1
    state["screen"] = "FIGHT_MENU"
    state["message"] = f"야생의 {wild_mon.name}이(가) 나타났다!"
    render()

MAP_DATA = [
    ["road","road","road","grass","grass","grass","road","road","road","road"],
    ["road","grass","grass","grass","road","road","road","grass","grass","road"],
    ["road","grass","wall","wall","road","grass","grass","grass","road","road"],
    ["road","grass","grass","road","road","road","wall","grass","grass","road"],
    ["road","road","grass","grass","grass","grass","road","road","grass","road"],
    ["road","grass","grass","wall","road","road","grass","grass","grass","road"],
    ["road","grass","road","road","grass","grass","road","wall","grass","road"],
    ["road","grass","grass","road","road","road","grass","grass","road","road"],
    ["road","road","grass","grass","grass","road","road","road","grass","road"],
    ["road","road","road","road","road","road","grass","road","road","road"],
]

player_pos = {"x": 0, "y": 0}
GRASS_RATE = 0.25
root_container = widgets.HBox(layout=widgets.Layout(justify_content="center"))

# ==========================
# 5. 턴 처리 및 액션 핸들러 (락 포함)
# ==========================
processing_lock = False

def handle_player_action(player_action):
    global processing_lock
    
    if processing_lock:
        return

    processing_lock = True
    try:
        with debug_out:
            debug_out.clear_output()

            if not ui_battle["active"] or ui_battle["player"] is None:
                print("배틀이 진행 중이 아닙니다.")
                return

            player = ui_battle["player"]
            enemy = ui_battle["enemy"]
            current_weather = ui_battle["weather"]

            if player.all_pokemon_fainted():
                ui_battle["active"] = False
                state["message"] = "사용할 포켓몬이 없습니다..."
                state["screen"] = "MAIN_MENU"
                render()
                return

            if enemy.current_pokemon.is_fainted and enemy.all_pokemon_fainted():
                ui_battle["active"] = False
                state["message"] = "상대 포켓몬이 이미 쓰러졌다!"
                state["screen"] = "MAIN_MENU"
                render()
                return

            if player_action["type"] == "run":
                if player.current_pokemon.is_fainted:
                    print("▶ 기절한 상태에서는 도망칠 수 없다!")
                else:
                    print("▶ 도망쳤다!")
                    ui_battle["active"] = False
                    state["screen"] = "MAIN_MENU"
                    state["message"] = "무사히 도망쳤다!"
                    render()
                return

            pmon = player.current_pokemon

            # ===== CASE A: 기절 시 강제 교체 (턴 소모 없음) =====
            if pmon.is_fainted:
                if player_action["type"] != "switch":
                    state["message"] = "기절한 포켓몬은 교체해야 합니다!"
                    render()
                    return

                idx = player_action["index"]
                if idx < 0 or idx >= len(player.team):
                    return
                new_mon = player.team[idx]

                if new_mon.is_fainted:
                    state["message"] = "그 포켓몬도 기절해 있습니다!"
                    render()
                    return
                
                print(f"▶ 돌아와, {pmon.name}!")
                player.current_pokemon = new_mon
                print(f"▶ 가라, {new_mon.name}!")
                
                state["message"] = f"{new_mon.name}(으)로 교체했다!"
                state["screen"] = "FIGHT_MENU"
                render()
                return

            # ===== CASE B: 정상 배틀 턴 =====
            player_choice = None
            
            if player_action["type"] == "attack":
                move_index = player_action["move_index"]
                if move_index < 0 or move_index >= len(pmon.moves):
                    return
                player_move = pmon.moves[move_index]
                player_choice = {
                    'action': 'attack',
                    'priority': player_move.get('priority', 0),
                    'move': player_move
                }
                
            elif player_action["type"] == "switch":
                idx = player_action["index"]
                if idx < 0 or idx >= len(player.team):
                    return
                new_mon = player.team[idx]

                if new_mon.is_fainted:
                    state["message"] = f"{new_mon.name}은(는) 기절해서 내보낼 수 없다!"
                    render()
                    return
                if new_mon == pmon:
                    state["message"] = f"{new_mon.name}은(는) 이미 배틀 중이다!"
                    render()
                    return
                
                player_choice = {'action': 'switch', 'priority': 6, 'index': idx}
                
            elif player_action["type"] == "item":
                item_name = player_action["item"]
                if item_name not in player.items or player.items[item_name] <= 0:
                    print(f"▶ {item_name}이(가) 없습니다!")
                    return
                player_choice = {'action': 'item', 'priority': 6, 'item': item_name}
            else:
                return

            enemy_choice = enemy.get_ai_choice()

            def get_speed(pokemon):
                s = pokemon.speed
                if pokemon.status == '마비':
                    s //= 4
                return s

            p_speed = get_speed(player.current_pokemon)
            o_speed = get_speed(enemy.current_pokemon)

            actions = [
                {'actor': player, 'choice': player_choice, 'speed': p_speed},
                {'actor': enemy,  'choice': enemy_choice,  'speed': o_speed},
            ]
            
            actions.sort(key=lambda x: (x['choice']['priority'], x['speed']), reverse=True)

            print(f"=== 턴 {ui_battle['turn']} ===")
            battle_ended = False

            for act in actions:
                if battle_ended:
                    break
                actor = act['actor']
                choice = act['choice']
                target = enemy if actor == player else player

                if actor.current_pokemon.is_fainted:
                    continue
                
                if choice['action'] == 'attack':
                    if not actor.current_pokemon.check_can_move():
                        continue
                    actor.current_pokemon.attack_opponent(target.current_pokemon, choice['move'], current_weather)
                    
                    if target.current_pokemon.is_fainted and target.all_pokemon_fainted():
                        msg = f"{actor.name} 승리!"
                        print(f"🎉 {msg} 🎉")
                        ui_battle["active"] = False
                        state["message"] = msg
                        state["screen"] = "MAIN_MENU"
                        battle_ended = True
                        render()
                        return
                
                elif choice['action'] == 'switch':
                    if actor == player:
                        success = actor.switch_pokemon(choice['index'])
                        if success:
                            state["message"] = f"{actor.current_pokemon.name}(으)로 교체했다!"
                        else:
                            print("▶ 교체 실패")
                
                elif choice['action'] == 'item':
                    if actor == player:
                        actor.use_item(choice['item'], actor.current_pokemon)
                        state["message"] = f"{actor.current_pokemon.name}에게 {choice['item']}을(를) 사용했다!"

            if not battle_ended:
                print("\n--- 턴 종료 (상태이상 데미지) ---")
                player.current_pokemon.apply_end_of_turn_status_damage()
                enemy.current_pokemon.apply_end_of_turn_status_damage()

                if player.all_pokemon_fainted():
                    print("패배했습니다...")
                    ui_battle["active"] = False
                    state["message"] = "패배했습니다..."
                    state["screen"] = "MAIN_MENU"
                    render()
                    return

                if player.current_pokemon.is_fainted:
                    state["message"] = f"{player.current_pokemon.name}이(가) 쓰러졌다! 다음 포켓몬을 선택하세요."
                    state["screen"] = "BATTLE_PARTY"
                else:
                    ui_battle["turn"] += 1
                    state["message"] = f"턴 {ui_battle['turn']-1} 종료. 무엇을 할까?"
                    state["screen"] = "FIGHT_MENU"
                
                render()

    finally:
        processing_lock = False

def handle_player_move(move_index):
    action = {"type": "attack", "move_index": move_index}
    handle_player_action(action)

# ==========================
# 5. UI 액션 핸들러
# ==========================
def handle_action(action):
    with debug_out:
        debug_out.clear_output()
    t = action["type"]
    
    if t == "OPEN_BAG":
        state["screen"] = "BAG_CATEGORY"
        state["message"] = "가방을 열었다."
    
    elif t == "BACK_TO_MAIN":
        if ui_battle["active"]:
            state["screen"] = "FIGHT_MENU"
            state["message"] = "무엇을 할까?"
        else:
            state["screen"] = "MAIN_MENU"
            state["message"] = "지우는 무엇을 할까?"

    elif t == "SELECT_ITEM":
        item = action["item"]
        if item["count"] <= 0:
            state["message"] = "아이템이 없습니다."
        
        elif ui_battle["active"]:
            handle_player_action({"type": "item", "item": item['name']})
        
        else:
            state["screen"] = "MAIN_MENU"
            state["message"] = f"{item['name']}을(를) 사용했다! (필드 효과 미구현)"

    elif t == "OPEN_PARTY":
        state["screen"] = "PARTY_SELECT"
        state["message"] = "포켓몬을 선택해 주십시오."

    elif t == "TRY_RUN":
        if not ui_battle["active"]:
            state["screen"] = "MAIN_MENU"
            state["message"] = "도망칠 곳이 없다..."
        else:
            handle_player_action({"type": "run"})

    elif t == "OPEN_BAG_CATEGORY":
        state["screen"] = "BAG_ITEMS"
        state["bag_category"] = action["category"]
        state["bag_page"] = 0

    elif t == "BACK_TO_BAG_CATEGORY":
        state["screen"] = "BAG_CATEGORY"

    elif t == "PAGE_LEFT":
        if state["bag_page"] > 0:
            state["bag_page"] -= 1

    elif t == "PAGE_RIGHT":
        cat = state["bag_category"]
        total = len(state["bag"][cat])
        max_page = (total - 1) // 6
        if state["bag_page"] < max_page:
            state["bag_page"] += 1

    elif t == "SELECT_PARTY_MEMBER":
        if ui_battle["player"] and ui_battle["active"]:
            mon = ui_battle["player"].team[action["index"]]
            state["screen"] = "MAIN_MENU"
            state["message"] = f"{mon.name}을(를) 선택했다! (더미)"
        else:
            mon = state["party"][action["index"]]
            state["screen"] = "MAIN_MENU"
            state["message"] = f"{mon['name']}을(를) 선택했다! (더미)"

    render()

def move_player(dx, dy):
    nx = player_pos["x"] + dx
    ny = player_pos["y"] + dy
    if not (0 <= nx < len(MAP_DATA[0]) and 0 <= ny < len(MAP_DATA)):
        return
    if MAP_DATA[ny][nx] == "wall":
        return
    player_pos["x"] = nx
    player_pos["y"] = ny
    tile = MAP_DATA[ny][nx]
    if tile == "grass" and random.random() < GRASS_RATE:
        start_ui_battle()
        return
    state["message"] = "어디로 갈까?"
    render()

# ==========================
# 6. 부분 렌더링 함수들
# ==========================
def render_map_panel():
    rows = []
    for y, row_data in enumerate(MAP_DATA):
        line = ""
        for x, tile in enumerate(row_data):
            if x == player_pos["x"] and y == player_pos["y"]:
                line += "🚶"
            elif tile == "road":
                line += "⬜"
            elif tile == "grass":
                line += "🌿"
            elif tile == "wall":
                line += "🧱"
        rows.append(widgets.HTML(f"<div style='font-size:20px; line-height:20px'>{line}</div>"))
    map_box = widgets.VBox(
        rows,
        layout=widgets.Layout(
            width=f"{PANEL_WIDTH - 20}px",
            height=f"{BATTLE_HEIGHT}px",
            align_items="flex-start"
        )
    )

    btn_layout = widgets.Layout(width="60px", height="60px")
    u = widgets.Button(description="▲", layout=btn_layout)
    d = widgets.Button(description="▼", layout=btn_layout)
    l = widgets.Button(description="◀", layout=btn_layout)
    r = widgets.Button(description="▶", layout=btn_layout)
    for b in (u, d, l, r):
        b.add_class("poke-btn")
        b.add_class("poke-green")
    u.on_click(lambda _: move_player(0, -1))
    d.on_click(lambda _: move_player(0, 1))
    l.on_click(lambda _: move_player(-1, 0))
    r.on_click(lambda _: move_player(1, 0))
    top_row = widgets.HBox([u], layout=widgets.Layout(justify_content="center"))
    bottom_row = widgets.HBox([l, d, r], layout=widgets.Layout(justify_content="center"))
    arrows = widgets.VBox([top_row, bottom_row], layout=widgets.Layout(align_items="center", margin="10px 180px 0 0"))
    return widgets.VBox([map_box, arrows], layout=widgets.Layout(width=f"{PANEL_WIDTH}px", align_items="center", justify_content="flex-start"))

def render_main_menu():
    def on_fight_click(_):
        start_ui_battle()
    fight_btn = widgets.Button(
        description="싸운다",
        layout=widgets.Layout(width="80%", height=f"{FIGHT_HEIGHT}px", margin="5px 0 15px 0")
    )
    fight_btn.add_class("poke-btn")
    fight_btn.add_class("poke-red")
    fight_btn.on_click(on_fight_click)

    if not ui_battle["active"]:
        fight_btn.disabled = True
    else:
        fight_btn.disabled = False

    bag_btn = widgets.Button(description="가방", layout=widgets.Layout(width=f"{BTN_W_SMALL}px", height=f"{BTN_H_SMALL}px"))
    bag_btn.add_class("poke-btn")
    bag_btn.add_class("poke-brown")

    run_btn = widgets.Button(description="도망간다", layout=widgets.Layout(width=f"{BTN_W_SMALL}px", height=f"{BTN_H_SMALL}px", margin="8px"))
    run_btn.add_class("poke-btn")
    run_btn.add_class("poke-blue")

    mon_btn = widgets.Button(description="포켓몬", layout=widgets.Layout(width=f"{BTN_W_SMALL}px", height=f"{BTN_H_SMALL}px"))
    mon_btn.add_class("poke-btn")
    mon_btn.add_class("poke-green")

    if not ui_battle["active"]:
        run_btn.disabled = True

    bag_btn.on_click(lambda _: handle_action({"type": "OPEN_BAG"}))
    run_btn.on_click(lambda _: handle_action({"type": "TRY_RUN"}))
    mon_btn.on_click(lambda _: handle_action({"type": "OPEN_PARTY"}))

    row = widgets.HBox([bag_btn, run_btn, mon_btn], layout=widgets.Layout(justify_content="center"))
    return widgets.VBox(
        [fight_btn, row],
        layout=widgets.Layout(
            align_items="center",
            width=f"{PANEL_WIDTH}px",
            height=f"{SCREEN_HEIGHT}px",
            justify_content="space-between"
        )
    )

def render_bag_category():
    top_area_height = SCREEN_HEIGHT - MSG_HEIGHT
    row_gap = 10
    btn_height = (top_area_height - row_gap) // 2 - 10

    def cat_btn(text, cat):
        b = widgets.Button(description=text, layout=widgets.Layout(width=f"{CAT_W}px", height=f"{btn_height}px"))
        b.add_class("poke-btn")
        b.add_class("poke-brown")
        b.on_click(lambda _: handle_action({"type": "OPEN_BAG_CATEGORY", "category": cat}))
        return b

    row1 = widgets.HBox(
        [cat_btn("HP/PP 회복", "HP_PP"), cat_btn("볼", "BALL")],
        layout=widgets.Layout(justify_content="center", margin=f"0 0 {row_gap}px 0")
    )
    row2 = widgets.HBox(
        [cat_btn("상태 회복", "STATUS"), cat_btn("배틀용", "BATTLE")],
        layout=widgets.Layout(justify_content="center")
    )

    top_box = widgets.VBox(
        [row1, row2],
        layout=widgets.Layout(width=f"{PANEL_WIDTH}px", height=f"{top_area_height}px")
    )

    msg_html = widgets.HTML(
        value=f"<div style='width:{PANEL_WIDTH - 135}px; height:{MSG_HEIGHT}px; "
              f"background:#e0e0e0; border:2px solid #000; box-sizing:border-box; "
              f"padding:8px; font-size:14px;'>{state['message']}</div>"
    )

    back_btn = widgets.Button(
        description="↲",
        layout=widgets.Layout(width="120px", height=f"{MSG_HEIGHT}px", margin="0 0 0 5px")
    )
    back_btn.add_class("poke-btn")
    back_btn.on_click(lambda _: handle_action({"type": "BACK_TO_MAIN"}))

    bottom_box = widgets.HBox(
        [msg_html, back_btn],
        layout=widgets.Layout(width=f"{PANEL_WIDTH}px", justify_content="flex-end")
    )

    return widgets.VBox(
        [top_box, bottom_box],
        layout=widgets.Layout(width=f"{PANEL_WIDTH}px", height=f"{SCREEN_HEIGHT}px")
    )

def render_bag_items():
    cat = state["bag_category"]
    items = state["bag"][cat]
    page = state["bag_page"]
    start = page * 6
    end = start + 6

    slot_btns = []
    for item in items[start:end]:
        b = widgets.Button(
            description=f"{item['name']} x{item['count']}",
            layout=widgets.Layout(width=f"{ITEM_W}px", height=f"{ITEM_H}px")
        )
        b.add_class("poke-btn")
        b.add_class("poke-brown")
        b.on_click(lambda _, i=item: handle_action({"type": "SELECT_ITEM", "item": i}))
        slot_btns.append(b)

    while len(slot_btns) < 6:
        b = widgets.Button(
            description="",
            disabled=True,
            layout=widgets.Layout(width=f"{ITEM_W}px", height=f"{ITEM_H}px")
        )
        b.add_class("poke-btn")
        slot_btns.append(b)

    grid = widgets.GridBox(
        children=slot_btns,
        layout=widgets.Layout(grid_template_columns=f"repeat(2, {ITEM_W+5}px)", grid_gap="5px")
    )

    left_btn = widgets.Button(description="<", layout=widgets.Layout(width=f"{NAV_W}px"))
    right_btn = widgets.Button(description=">", layout=widgets.Layout(width=f"{NAV_W}px"))
    left_btn.add_class("poke-btn")
    right_btn.add_class("poke-btn")
    left_btn.on_click(lambda _: handle_action({"type": "PAGE_LEFT"}))
    right_btn.on_click(lambda _: handle_action({"type": "PAGE_RIGHT"}))

    page_nav = widgets.HBox(
        [left_btn, widgets.Label(f"{page+1}"), right_btn],
        layout=widgets.Layout(justify_content="center")
    )

    back_btn = widgets.Button(
        description="뒤로",
        layout=widgets.Layout(width=f"{CAT_W}px", height=f"{BACK_H}px")
    )
    back_btn.add_class("poke-btn")
    back_btn.add_class("poke-brown")
    back_btn.on_click(lambda _: handle_action({"type": "BACK_TO_BAG_CATEGORY"}))

    return widgets.VBox(
        [grid, page_nav, back_btn],
        layout=widgets.Layout(
            width=f"{PANEL_WIDTH}px",
            height=f"{SCREEN_HEIGHT}px",
            align_items="center",
            justify_content="space-between"
        )
    )

def render_party():
    top_area_height = SCREEN_HEIGHT - MSG_HEIGHT
    row_gap = 5
    party_btn_height = int(6 * (top_area_height - 2 * row_gap) / 19)
    if party_btn_height < 30:
        party_btn_height = 30
    col_offset = int(party_btn_height / 6)
    party_btn_width = int((PANEL_WIDTH - 40) / 2)
    
    left_indices = [0, 2, 4]
    right_indices = [1, 3, 5]

    def make_party_button(idx, extra_top_margin=0):
        if ui_battle["player"] and idx < len(ui_battle["player"].team):
            p = ui_battle["player"].team[idx]
            status_text = ""
            if p.is_fainted:
                status_text = " [기절]"
            elif p.status:
                status_text = f" [{p.status}]"
            desc = f"{p.name} Lv.{p.level} ({p.current_hp}/{p.max_hp}){status_text}"
            b = widgets.Button(
                description=desc,
                layout=widgets.Layout(
                    width=f"{party_btn_width}px",
                    height=f"{party_btn_height}px",
                    margin=f"{extra_top_margin}px 0 0 0"
                )
            )
            b.add_class("poke-btn")
            if p.is_fainted:
                b.add_class("poke-red")
            else:
                b.add_class("poke-green")
            b.on_click(lambda _, i=idx: handle_action({"type": "SELECT_PARTY_MEMBER", "index": i}))
            return b
        else:
            b = widgets.Button(
                description="-",
                disabled=True,
                layout=widgets.Layout(
                    width=f"{party_btn_width}px",
                    height=f"{party_btn_height}px",
                    margin=f"{extra_top_margin}px 0 0 0"
                )
            )
            b.add_class("poke-btn")
            return b

    left_buttons = []
    for j, idx in enumerate(left_indices):
        top_margin = 0 if j == 0 else row_gap
        left_buttons.append(make_party_button(idx, extra_top_margin=top_margin))

    left_col = widgets.VBox(
        left_buttons,
        layout=widgets.Layout(
            width=f"{party_btn_width}px",
            height=f"{top_area_height}px",
            align_items="flex-start",
            justify_content="flex-start",
            margin="0 5px 0 0"
        )
    )

    right_buttons = []
    for j, idx in enumerate(right_indices):
        top_margin = col_offset if j == 0 else row_gap
        right_buttons.append(make_party_button(idx, extra_top_margin=top_margin))

    right_col = widgets.VBox(
        right_buttons,
        layout=widgets.Layout(
            width=f"{party_btn_width}px",
            height=f"{top_area_height}px",
            align_items="flex-start",
            justify_content="flex-start",
            margin="0 0 0 5px"
        )
    )

    top_box = widgets.HBox(
        [left_col, right_col],
        layout=widgets.Layout(
            width=f"{PANEL_WIDTH}px",
            height=f"{top_area_height}px",
            justify_content="center"
        )
    )
    
    msg_html = widgets.HTML(
        value=f"<div style='width:{PANEL_WIDTH - 135}px; height:{MSG_HEIGHT}px; "
              f"background:#f5f0dc; border:2px solid #000; box-sizing:border-box; "
              f"padding:8px; font-size:14px;'>{state['message']}</div>"
    )
    
    back_btn = widgets.Button(
        description="뒤로",
        layout=widgets.Layout(width="120px", height=f"{MSG_HEIGHT}px", margin="0 0 0 5px")
    )
    back_btn.add_class("poke-btn")
    back_btn.add_class("poke-green")
    back_btn.on_click(lambda _: handle_action({"type": "BACK_TO_MAIN"}))
    
    bottom_box = widgets.HBox(
        [msg_html, back_btn],
        layout=widgets.Layout(width=f"{PANEL_WIDTH}px", justify_content="flex-end")
    )
    
    return widgets.VBox(
        [top_box, bottom_box],
        layout=widgets.Layout(width=f"{PANEL_WIDTH}px", height=f"{SCREEN_HEIGHT}px")
    )

def render_battle_party():
    player = ui_battle["player"]
    team = player.team
    btn_width = (PANEL_WIDTH - 30) // 2
    btn_height = "60px"
    party_btns = []

    for i in range(6):
        if i < len(team):
            mon = team[i]
            status_str = ""
            if mon.is_fainted:
                status_str = " [기절]"
            elif mon.status:
                status_str = f" [{mon.status}]"
            desc = f"{mon.name} (Lv.{mon.level}) {mon.current_hp}/{mon.max_hp}{status_str}"

            btn = widgets.Button(
                description=desc,
                layout=widgets.Layout(width=f"{btn_width}px", height=btn_height)
            )
            btn.add_class("poke-btn")
            if mon == player.current_pokemon:
                btn.add_class("poke-green")
                btn.disabled = True
            elif mon.is_fainted:
                btn.add_class("poke-red")
                btn.disabled = True
            else:
                btn.add_class("poke-blue")

            btn.on_click(lambda _, idx=i: handle_player_action({"type": "switch", "index": idx}))
            party_btns.append(btn)
        else:
            btn = widgets.Button(
                description="",
                disabled=True,
                layout=widgets.Layout(width=f"{btn_width}px", height=btn_height)
            )
            btn.add_class("poke-btn")
            party_btns.append(btn)

    party_grid = widgets.GridBox(
        children=party_btns,
        layout=widgets.Layout(
            grid_template_columns=f"repeat(2, {btn_width}px)",
            grid_gap="5px",
            justify_content="center",
            margin="10px 0 10px 0"
        )
    )

    msg_html = widgets.HTML(
        value=f"<div style='width:{PANEL_WIDTH - 135}px; height:{MSG_HEIGHT}px; "
              f"background:#f5f0dc; border:2px solid #000; box-sizing:border-box; "
              f"padding:8px; font-size:14px;'>교체할 포켓몬을 선택하세요.<br>{state['message']}</div>"
    )

    back_btn = widgets.Button(
        description="뒤로",
        layout=widgets.Layout(width="120px", height=f"{MSG_HEIGHT}px", margin="0 0 0 5px")
    )
    back_btn.add_class("poke-btn")
    back_btn.add_class("poke-green")
    back_btn.on_click(lambda _: (state.update({"screen": "FIGHT_MENU"}), render()))

    bottom_box = widgets.HBox(
        [msg_html, back_btn],
        layout=widgets.Layout(width=f"{PANEL_WIDTH}px", justify_content="flex-end")
    )

    return widgets.VBox(
        [party_grid, bottom_box],
        layout=widgets.Layout(
            width=f"{PANEL_WIDTH}px",
            height=f"{SCREEN_HEIGHT}px",
            justify_content="space-between"
        )
    )

def render_battle_bag():
    player = ui_battle["player"]
    items = player.items
    item_buttons = []

    if not items:
        lbl = widgets.Label("사용할 수 있는 아이템이 없습니다.")
        item_buttons.append(lbl)
    else:
        for name, count in items.items():
            desc = f"{name} x{count}"
            btn = widgets.Button(
                description=desc,
                layout=widgets.Layout(
                    width=f"{PANEL_WIDTH - 40}px",
                    height="45px",
                    margin="3px 0 3px 0"
                )
            )
            btn.add_class("poke-btn")
            btn.add_class("poke-brown")

            def on_click(_, item_name=name):
                handle_player_action({"type": "item", "item": item_name})
            btn.on_click(on_click)

            item_buttons.append(btn)

    top_box = widgets.VBox(
        item_buttons,
        layout=widgets.Layout(
            width=f"{PANEL_WIDTH}px",
            height=f"{SCREEN_HEIGHT - MSG_HEIGHT}px",
            overflow_y="auto"
        )
    )

    msg_html = widgets.HTML(
        value=f"<div style='width:{PANEL_WIDTH - 135}px; height:{MSG_HEIGHT}px; "
              f"background:#e0e0e0; border:2px solid #000; box-sizing:border-box; "
              f"padding:8px; font-size:14px;'>사용할 아이템을 선택하세요.<br>{state['message']}</div>"
    )

    back_btn = widgets.Button(
        description="뒤로",
        layout=widgets.Layout(width="120px", height=f"{MSG_HEIGHT}px", margin="0 0 0 5px")
    )
    back_btn.add_class("poke-btn")
    back_btn.on_click(lambda _: (state.update({"screen": "FIGHT_MENU"}), render()))

    bottom_box = widgets.HBox(
        [msg_html, back_btn],
        layout=widgets.Layout(width=f"{PANEL_WIDTH}px", justify_content="flex-end")
    )

    return widgets.VBox(
        [top_box, bottom_box],
        layout=widgets.Layout(width=f"{PANEL_WIDTH}px", height=f"{SCREEN_HEIGHT}px")
    )

def render_fight_menu():
    p = ui_battle["player"].current_pokemon
    moves = p.moves
    move_btns = []
    BTN_H_SKILL = "75px"

    for i in range(4):
        if i < len(moves):
            m = moves[i]
            desc = f"{m['name']} ({m['type']})"
            b = widgets.Button(
                description=desc,
                layout=widgets.Layout(
                    width=f"{(PANEL_WIDTH - 30) // 2}px",
                    height=BTN_H_SKILL
                )
            )
            b.add_class("poke-btn")
            b.add_class("poke-red")
            b.on_click(lambda _, idx=i: handle_player_move(idx))
            move_btns.append(b)
        else:
            b = widgets.Button(
                disabled=True,
                layout=widgets.Layout(
                    width=f"{(PANEL_WIDTH - 30) // 2}px",
                    height=BTN_H_SKILL
                )
            )
            b.add_class("poke-btn")
            b.add_class("poke-red")
            move_btns.append(b)

    moves_grid = widgets.GridBox(
        children=move_btns,
        layout=widgets.Layout(
            grid_template_columns=f"repeat(2, {(PANEL_WIDTH - 20) // 2}px)",
            grid_gap="5px",
            margin="0 0 15px 0"
        )
    )
    
    bag_btn = widgets.Button(
        description="가방",
        layout=widgets.Layout(width=f"{BTN_W_SMALL}px", height=f"{BTN_H_SMALL}px")
    )
    bag_btn.add_class("poke-btn")
    bag_btn.add_class("poke-brown")
    bag_btn.on_click(lambda _: handle_action({"type": "OPEN_BAG"}))
    
    mon_btn = widgets.Button(
        description="포켓몬",
        layout=widgets.Layout(width=f"{BTN_W_SMALL}px", height=f"{BTN_H_SMALL}px")
    )
    mon_btn.add_class("poke-btn")
    mon_btn.add_class("poke-green")
    mon_btn.on_click(lambda _: (state.update({"screen": "BATTLE_PARTY"}), render()))
    
    run_btn = widgets.Button(
        description="도망간다",
        layout=widgets.Layout(width=f"{BTN_W_SMALL}px", height=f"{BTN_H_SMALL}px", margin="8px")
    )
    run_btn.add_class("poke-btn")
    run_btn.add_class("poke-blue")
    run_btn.on_click(lambda _: handle_player_action({"type": "run"}))
    
    bottom_menu = widgets.HBox(
        [bag_btn, run_btn, mon_btn],
        layout=widgets.Layout(justify_content="center")
    )
    return widgets.VBox(
        [moves_grid, bottom_menu],
        layout=widgets.Layout(
            width=f"{PANEL_WIDTH}px",
            height=f"{SCREEN_HEIGHT}px",
            justify_content="space-between",
            align_items="center"
        )
    )
def render():
    # ----- 왼쪽 영역: 지도 or 배틀 화면 -----
    if ui_battle["active"]:
        p = ui_battle["player"].current_pokemon
        e = ui_battle["enemy"].current_pokemon

        # 💥 피격 흔들림용 CSS 클래스 선택
        p_shake = "shake" if getattr(p, "just_took_damage", False) else ""
        e_shake = "shake" if getattr(e, "just_took_damage", False) else ""

        # HP 바 계산
        e_per = (e.current_hp / e.max_hp) * 100 if e.max_hp > 0 else 0
        e_color = "#4caf50"
        if e_per <= 20:
            e_color = "#f44336"
        elif e_per <= 50:
            e_color = "#ffeb3b"
        
        p_per = (p.current_hp / p.max_hp) * 100 if p.max_hp > 0 else 0
        p_color = "#4caf50"
        if p_per <= 20:
            p_color = "#f44336"
        elif p_per <= 50:
            p_color = "#ffeb3b"

        player_sprite = f"/files/PNG/{p.name}_뒤.png"
        enemy_sprite  = f"/files/PNG/{e.name}_앞.png"

        battle_html = f"""
        <div style="
            width:{PANEL_WIDTH}px;
            height:{BATTLE_HEIGHT}px;
            background-image:url('/files/PNG/전투배경.png');
            background-size: cover;
            background-repeat: no-repeat;
            background-position: center;
            box-sizing:border-box;
            padding:15px;
            color:#000;
            font-size:14px;
            position:relative;
            overflow:hidden;
        ">
            <!-- 적 상태 박스 -->
            <div style="
                position:absolute;
                left:20px;
                top:10px;
                padding:6px 10px;
                border:2px solid #000;
                border-radius:8px;
                background:rgba(255,255,255,0.8);
                font-size:13px;
                z-index:10;
            ">
                <b>{e.name}</b> Lv.{e.level}<br>
                <div style="display:flex; align-items:center; margin-top:2px;">
                    <span style="font-size:10px; font-weight:bold; margin-right:4px;">HP</span>
                    <div style="
                        width:100px;
                        height:8px;
                        background:#555;
                        border-radius:4px;
                        border:1px solid #222;
                        overflow:hidden;
                    ">
                        <div style="
                            width:{e_per}%;
                            height:100%;
                            background:{e_color};
                        "></div>
                    </div>
                </div>
            </div>

            <!-- 아군 상태 박스 -->
            <div style="
                position:absolute;
                right:20px;
                bottom:10px;
                padding:6px 10px;
                border:2px solid #000;
                border-radius:8px;
                background:rgba(255,255,255,0.8);
                font-size:13px;
                z-index:10;
            ">
                <b>{p.name}</b> Lv.{p.level}<br>
                <div style="display:flex; align-items:center; margin-top:2px; margin-bottom:2px;">
                    <span style="font-size:10px; font-weight:bold; margin-right:4px;">HP</span>
                    <div style="
                        width:100px;
                        height:8px;
                        background:#555;
                        border-radius:4px;
                        border:1px solid #222;
                        overflow:hidden;
                    ">
                        <div style="
                            width:{p_per}%;
                            height:100%;
                            background:{p_color};
                        "></div>
                    </div>
                </div>
                <div style="text-align:right; font-size:12px;">
                    {p.current_hp} / {p.max_hp}
                </div>
            </div>

            <!-- 적 포켓몬 스프라이트 (피격 시 흔들림) -->
            <img src="{enemy_sprite}"
                 class="{e_shake}"
                 style="
                    position:absolute;
                    right:60px;
                    top:30px;
                    width:120px;
                    image-rendering:pixelated;
                    z-index:1;
                 "
                 onerror="this.style.display='none'">

            <!-- 아군 포켓몬 스프라이트 (피격 시 흔들림) -->
            <img src="{player_sprite}"
                 class="{p_shake}"
                 style="
                    position:absolute;
                    left:40px;
                    bottom:-30px;
                    width:120px;
                    image-rendering:pixelated;
                    z-index:1;
                 "
                 onerror="this.style.display='none'">
        </div>
        """

        battle_box = widgets.HTML(value=battle_html)
        msg_box = widgets.HTML(
            value=f"<div style='width:{PANEL_WIDTH}px; height:{MSG_HEIGHT}px; "
                  f"background:#fff; border:2px solid #000; box-sizing:border-box; "
                  f"padding:10px; font-size:15px;'>{state['message']}</div>"
        )
        left_panel = widgets.VBox(
            [battle_box, msg_box],
            layout=widgets.Layout(width=f"{PANEL_WIDTH}px")
        )

        # 💥 한 번 흔들린 뒤에는 플래그 초기화
        if hasattr(p, "just_took_damage"):
            p.just_took_damage = False
        if hasattr(e, "just_took_damage"):
            e.just_took_damage = False

    else:
        left_panel = render_map_panel()

    # ----- 오른쪽 패널: 화면 상태에 따라 전환 -----
    if state["screen"] == "MAIN_MENU":
        right_panel = render_main_menu()
    elif state["screen"] == "BAG_CATEGORY":
        right_panel = render_bag_category()
    elif state["screen"] == "BAG_ITEMS":
        right_panel = render_bag_items()
    elif state["screen"] == "PARTY_SELECT":
        right_panel = render_party()
    elif state["screen"] == "FIGHT_MENU":
        right_panel = render_fight_menu()
    elif state["screen"] == "BATTLE_BAG":
        right_panel = render_battle_bag()
    elif state["screen"] == "BATTLE_PARTY":
        right_panel = render_battle_party()
    else:
        right_panel = widgets.Label("오류 발생")

    # 오른쪽 상단 포켓몬볼 헤더
    header_html = f"""
    <div style="
        width:{PANEL_WIDTH}px;
        height:35px;
        box-sizing:border-box;
        padding:5px 10px;
        display:flex;
        align-items:flex-start;
        gap:4px;
    ">
        <img src="/files/PNG/포켓몬볼.1.png"
             style="height:100%; width:auto; image-rendering:pixelated;"
             onerror="this.style.display='none'">
        <img src="/files/PNG/포켓몬볼.1.png"
             style="height:100%; width:auto; image-rendering:pixelated;"
             onerror="this.style.display='none'">
        <img src="/files/PNG/포켓몬볼.1.png"
             style="height:100%; width:auto; image-rendering:pixelated;"
             onerror="this.style.display='none'">
        <img src="/files/PNG/포켓몬볼.1.png"
             style="height:100%; width:auto; image-rendering:pixelated;"
             onerror="this.style.display='none'">
        <img src="/files/PNG/포켓몬볼.1.png"
             style="height:100%; width:auto; image-rendering:pixelated;"
             onerror="this.style.display='none'">
    </div>
    """
    header_box = widgets.HTML(value=header_html)

    right = widgets.VBox(
        [header_box, right_panel],
        layout=widgets.Layout(width=f"{PANEL_WIDTH}px")
    )

    root_container.children = [left_panel, right]


# ==== 초기화 & 첫 렌더 ====
init_ui_battle()
display(root_container, debug_out)
render()
```
